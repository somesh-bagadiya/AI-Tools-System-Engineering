## Dealing with missing values

I am removing all the columns with more than 30% missing values, and then replacing missing values with mean for numerical columns and for categorical columns I am doing pd.get_dummies. This increased the number of columns from 79 to 280.

```python
missing_percentage = (train.isnull().sum() / len(train)) * 100

# Identify columns with missing values exceeding 50%
columns_to_drop = missing_percentage[missing_percentage > 30].index

# Remove columns with missing values exceeding 50%
train_cleaned = train.drop(columns=columns_to_drop)

# Check the resulting dataset
print("Original Shape:", train.shape)
print("Shape after removing columns with > 50% missing values:", train_cleaned.shape)
```

```python
train = train_cleaned
```

```python
columns_with_missing_values = train.columns[train.isnull().any()].tolist()

# Fill missing values based on the data type of the column
for column in columns_with_missing_values:
    if train[column].dtype == 'object':
        # For categorical columns, fill missing values with the mode
        train[column].fillna(train[column].mode()[0], inplace=True)
    else:
        # For numerical columns, fill missing values with the mean
        train[column].fillna(train[column].mean(), inplace=True)

train = pd.get_dummies(train)    
    
# Check if there are still any missing values
print("Remaining missing values:")
print(train.isnull().sum())
```

```python
train.drop(columns="patient_id", inplace=True)
train.head()
```

```python
train.shape
```

## Scaling the data

Before scaling I am removing the target column so that it dose not get scaled while scaling the whole dataset. I am using the standard scalar from the sklearn library to scale the dataset.

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target variable
X = train.drop(columns=['DiagPeriodL90D'])
y = train['DiagPeriodL90D']

# Apply feature scaling using Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Check the shape and first few rows of the scaled features
print("Shape of scaled features:", X_scaled.shape)
print(pd.DataFrame(X_scaled, columns=X.columns).head())

X = X_scaled
```

## Dimensionality Reduction

I am doing the PCA to reduce the number of features I am using to train the models. Based on the elbow method to decide number of components, I have decided to use 20 components to train the model.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Fit PCA
pca = PCA()
pca.fit(X)

# Plot explained variance ratio
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='-')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio vs. Number of Components')
plt.grid(True)
plt.show()

```

```python
pca = PCA(n_components=20)
X_pca = pca.fit_transform(X)

print("Shape of reduced features:", X_pca.shape)
```

## Training Model

I am using logistic regression, random forest and support vector classifier to predict the results of the model.

```python
# Step 1: Model Selection
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Initialize models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC()
}

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

for name, model in models.items():
    model.fit(X_train, y_train)

from sklearn.metrics import classification_report

for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"Model: {name}")
    print(classification_report(y_test, y_pred))
    print("="*50)
```
